# Composition

```python
notes = ['C', 'C#/Db', 'D', 'D#/Eb', 'E', 'F', 'F#/Gb', 'G', 'G#/Ab', 'A', 'A#/Bb', 'B']

interv_maj = pd.DataFrame([('I  ', '2 M, 3 M, 4 G, 5 G, 6 M, 7 M'),
        ('ii ', '2 M, 3 m, 4 G, 5 G, 6 M, 7 m'),
        ('iii', '2 m, 3 m, 4 G, 5 G, 6 M, 7 m'),
        ('IV ', '2 M, 3 m, 4 A, 5 G, 6 M, 7 M'),
        ('V  ', '2 M, 3 M, 4 G, 5 G, 6 M, 7 m'),
        ('vi ', '2 M, 3 m, 4 G, 5 G, 6 m, 7 m'),
        ('ø  ', '2 m, 3 m, 4 G, 5 d, 6 m, 7 m')])

interv_minor = pd.DataFrame([('i  ', '2 M, 3 m, 4 G, 5 G, 6 m, 7 m'),
    ('ii ', '2 m, 3 m, 4 G, 5 d, 6 m, 7 m'),
    ('III', '2 M, 3 M, 4 G, 5 G, 6 M, 7 M'),
    ('iv ', '2 M, 3 m, 4 G, 5 G, 6 M, 7 m'),
    ('v  ', '2 m, 3 m, 4 G, 5 G, 6 M, 7 m'),
    ('VI ', '2 M, 3 m, 4 A, 5 G, 6 M, 7 M'),
    ('VII ', '2 M, 3 M, 4 G, 5 G, 6 M, 7 m')])

alt_diesis = pd.DataFrame([('C Major', 0),
    ('G Major', 1),
    ('D Major', 2),
    ('A Major', 3),
    ('E Major', 4),
    ('B Major', 5),
    ('F# Major', 6),
    ('C# Major', 7)])

alt_bemolle = pd.DataFrame([('C Major', 0),
    ('F Major', 1),
    ('Bb Major', 2),
    ('Eb Major', 3),
    ('Ab Major', 4),
    ('Db Major', 5),
    ('Gb Major', 6),
    ('Cb Major', 7)])

def scale(root, mode_id):
    for i in range(len(notes)):
            if root == notes[i]:
                if mode_id == 0:
                    return pd.DataFrame([notes[i%len(notes)],
                        notes[(i+2)%len(notes)],
                        notes[(i+4)%len(notes)],
                        notes[(i+5)%len(notes)],
                        notes[(i+7)%len(notes)],
                        notes[(i+9)%len(notes)],
                        notes[(i+11)%len(notes)]], columns = ['Notes'])
                else:
                    return pd.DataFrame([notes[i%len(notes)],
                        notes[(i+2)%len(notes)],
                        notes[(i+3)%len(notes)],
                        notes[(i+5)%len(notes)],
                        notes[(i+7)%len(notes)],
                        notes[(i+9)%len(notes)],
                        notes[(i+10)%len(notes)]], columns = ['Notes'])

def chords(root, mode_id):
    for i in range(len(notes)):
        if root == notes[i]:
            if mode_id == 0:
                return pd.DataFrame([('I', notes[i]+' MAJ 7'),
                    ('ii', notes[(i+2)%12]+' min 7'),
                    ('iii', notes[(i+4)%12]+' min 7'),
                    ('IV', notes[(i+5)%12]+' MAJ 7'),
                    ('V', notes[(i+7)%12]+' 7'),
                    ('vi', notes[(i+9)%12]+' min 7'),
                    ('vii ø', notes[(i+11)%12]+' semidim')], columns = ['Grade', 'Chord'])   
            else:
                return  pd.DataFrame([('i', notes[i]+' min 7'),
                    ('ii ø', notes[(i+2)%12]+' semidim'),
                    ('III', notes[(i+3)%12]+' MAJ 7'),
                    ('iv', notes[(i+5)%12]+' min 7'),
                    ('v', notes[(i+7)%12]+' min 7'),
                    ('VI', notes[(i+8)%12]+' MAJ 7'),
                    ('VII', notes[(i+10)%12]+' 7')], columns = ['Grade', 'Chord'])

def intervals(mode):
    return interv_maj if mode_id == 0 else interv_minor

def find_key(alterazioni, n_alter):
    if alterazioni == 0: use = alt_diesis
    else: use = alt_bemolle
    for i in use[1]:
        if n_alter == use[1][i]:
            return use[0][i]
        
def major_to_minor(tonic):
    return notes[(tonic+9)%12]

def minor_to_major(tonic):
    return notes[(tonic+3)%12]
```

```python
for i in range(len(notes)):
    print(f'Key of {notes[i]}')
    print(f'{notes[i]} Major')
    display(scale(notes[i], 0).set_index('Notes').T)
    print(f'The relative minor is {major_to_minor(i)} minor')
    print(f'{notes[i]} major')
    display(chords(notes[i], 0))
    print(f'\n{notes[i]} minor')
    display(scale(notes[i], 1).T)
    print(f'The relative major is {minor_to_major(i)} major')
    display(chords(notes[i], 1))
```

Create a .txt file with all the possible keys and chords.

```python
with open('out.txt', 'w') as f:

    f.write(f'Composition Guide\n\n\n') # Title

    for i in range(len(notes)):
        f.write(f'Key of {notes[i]}\n\n')
        f.write(f'{notes[i]} Major\n')
        f.write(f'{scale(notes[i], 0).T.to_string(header=False)}\n')
        f.write(f'The relative minor scale is {major_to_minor(i)} minor\n\n')
        f.write(f'{notes[i]} Major\n')
        f.write(f'{chords(notes[i], 0).to_string(index=False)}\n\n')
        f.write(f'{notes[i]} Minor\n')
        f.write(f'{scale(notes[i], 1).T.to_string(header=False)}\n')
        f.write(f'The relative major scale is {minor_to_major(i)} major\n\n')
        f.write(f'{chords(notes[i], 1).to_string(index=False)}\n\n\n\n')
```

```python
pdf = FPDF() # save FPDF() class into a variable pdf

pdf.add_page() # add a page

f = open("out.txt", "r")

# insert the texts in pdf
pdf.set_font("Arial", size = 20)
pdf.cell(200, 10, txt = f.readline(), ln = 1, align = 'C') # change the title formatting

pdf.set_font("Courier", size = 16) # set style and size of font that you want in the pdf
for x in f.readlines()[1:]:
    pdf.cell(200, 10, txt = x, ln = 1, align = 'L')

pdf.output("Guida.pdf") # save the pdf with name .pdf
```
